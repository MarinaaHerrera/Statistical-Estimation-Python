# Hypothesis Testing

```python
# Set up the environment

import pandas as pd
import numpy as np
from scipy import stats
```

## Hilltop Coffee

```python
# --- 1. Define Hypotheses ---
# Null Hypothesis (H0): μ >= 3
# Alternative Hypothesis (Ha): μ < 3
# This is a one-tailed test.

# Hypothesized population mean (μ0)
hypothesized_mean = 3

# Significance level (alpha) - FTC wants to be 99% confidence
alpha = 0.01
```

```python
# --- 2. Sample Data ---
# read the csv file into a dataframe
df=pd.read_csv("C://Users//marin//Downloads//07_coffee (1).csv")

df.head()
```

```python
# --- 3. Perform the One-Sample T-Test ---
# The stats.ttest_1samp function returns the t-statistic and the two-tailed p-value
t_stat, p_value = stats.ttest_1samp(df['Weight'], popmean=hypothesized_mean)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value/2:.4f}") # Calculate the one-tailed p-value
```

 --- 4. Make a Decision ---

Since p_value = 0.0039, which is lower than 0.01. We can reject the null hypothesis based on the sample data. There is sufficient evidence to conclude that the filling weight of Hilltop Coffee cans is less than 3 lbs.

## Pine Creek Golf Course

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
# --- 1. Define Hypotheses ---
# Null Hypothesis (H0): p-hat <= 0.2
# Alternative Hypothesis (Ha): p-hat > 0.2
# This is a one-tailed test.

# Hypothesized population proportion (p-bar)
p_bar = 0.20

# Significance level (alpha) 
alpha = 0.05

# Alternative hypothesis: 'two-sided', 'smaller', or 'larger'
# 'two-sided': tests if the true proportion is not equal to the value
# 'larger': tests if the true proportion is greater than the value
# 'smaller': tests if the true proportion is less than the value
tailed = 'larger' 
```

```python
# --- 2. Sample Data ---
# read the csv file into a dataframe
df = pd.read_csv("C://Users//marin//Downloads//07_womengolf (1).csv")

df.head()
```

```python
# Number of female in the sample
female_count = (df['Golfer'] == 'Female').sum() 

# Total number of observations in the sample
nobs = len(df)
```

```python
# ----- 3. Perform the Z-test -----
z_stat, p_value = proportions_ztest(female_count, nobs, p_bar, alternative = tailed)

# ----- Print the results -----
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")
```

The z-stat is a little smaller than our manual calculation because the proportions_ztest function applies a continuity correction by default (unless specified otherwise). This is a technique used to improve the accuracy of a normal approximation to a binomial distribution. Manual calculation often omits this step.

-- 4. Make a Decision ---

Since p_value = 0.0105, which is smaller than alpha = 0.05. We can reject the null hypothesis based on the sample data. There is sufficient evidence to conclude that the proportion of women golfers at Pine Creek Golf Course has increased.

### Exercise: Malpractice Suits

One of the reasons health-care costs have been rising rapidly in recent years is the increasing cost of malpractice insurance for physicians. Also, fear of being sued causes doctors to run more precautionary tests (possibly unnecessary) just to make sure they are not guilty of missing something. These precautionary tests also add to health-care costs. Data in the file _07_lawsuit.csv_ are consistent with findings in a Reader’s Digest article and can be used to estimate the proportion of physicians over the age of 55 who have been sued at least once.

1.	Formulate hypotheses that can be used to see if these data can support a finding that more than half of physicians over the age of 55 have been sued at least once.
1.	Use the data file to compute the sample proportion of physicians over the age of 55 who have been sued at least once. What is the p value for your hypothesis test?
1.	At α = 0.01, what is your conclusion?


```python
df = pd.read_csv("C://Users//marin//Downloads//07_lawsuit.csv")
df
```

```python
import pandas as pd
import numpy as np
import math

#1
print("1. Ho: p <= 0.5")
print("   Ha: p > 0.5")

#2
df = pd.read_csv(r"C:\Users\marin\Downloads\07_lawsuit.csv")

sued_count = (df['Have you been sued by a patient?'] == 'Yes').sum()
nobs = len(df)
p_hat = sued_count / nobs

z_stat = (p_hat - 0.5) / math.sqrt(0.5 * 0.5 / nobs)
p_value = 1 - (0.5 * (1 + math.erf(z_stat / math.sqrt(2))))

print("2. P-value:", round(p_value, 4))

#3
print("3. Since the p-value is less than 0.01, we reject the null hypothesis. There is sufficient evidence to conclude that more than half of physicians over the age of 55 have been sued at least once.")
```

### Exercise: Social Networks

CEOs who belong to a popular business-oriented social networking service have an average of 930 connections. Do other members have fewer connections than CEOs? The number of connections for a random sample of 7,515 members who are not CEOs is provided in the file _07_socialnetwork.csv_. Using this sample, test the hypothesis that other members have fewer connections than CEOs at   Discuss the practical significance of the results.

```python
#Ho: μ >= 930
#Ha: μ < 930

df = pd.read_csv(r"C:\Users\marin\Downloads\07_socialnetwork.csv")

xbar = df['Connections'].mean()
s = df['Connections'].std(ddof=1)   # sample standard deviation
n = len(df)

t_stat = (xbar - 930) / (s / np.sqrt(n))
p_value = 0.5 * (1 + math.erf(t_stat / math.sqrt(2)))

print("Sample mean:", round(xbar, 2))
print("T-statistic:", round(t_stat, 4))
print("P-value (one-tailed):", round(p_value, 4))
print("We reject the hypothesis since p < 0.05. There is evidence that non-CEOs have fewer connections.")
```
