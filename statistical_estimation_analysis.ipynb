# Statistical Estimation

```python
# Import the necessary libraries

import numpy as np
from scipy import stats
```

### Mean Estimation

```python
# Sample data
data = [85, 92, 78, 88, 95, 80, 70, 90, 83]
confidence_level = 0.95

# Calculate sample mean (point estimation) and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # ddof=1 for sample standard deviation

# Get the sample size
n = len(data)

# Calculate the standard error of the mean (SEM)
sem = sample_std / np.sqrt(n)

# Calculate the confidence interval
confidence_interval = stats.t.interval(confidence_level, n - 1, loc=sample_mean, scale=sem)

print(f"Sample Mean (Point Estimate of Population Mean): {sample_mean:.2f}")
print(f"95% Confidence Interval for the Population Mean: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
```

### Proportion Estimation

```python
# Sample data
num_successes = 60
total_trials = 200

# Desired confidence level
confidence_level = 0.95

# Calculate the sample proportion (p-hat)
p_hat = num_successes / total_trials
print(f"Sample Proportion (p-hat): {p_hat:.4f}")

# Calculate the standard error of the proportion
# This is the standard deviation of the sampling distribution of p-hat
standard_error = np.sqrt((p_hat * (1 - p_hat)) / total_trials)

# Find the critical Z-score for the desired confidence level
# For a 95% confidence interval, we need the Z-score that leaves 2.5% in each tail
# The `ppf` (percent point function) is the inverse of the CDF
alpha = 1 - confidence_level
z_critical = stats.norm.ppf(1 - alpha / 2) # e.g., for 0.95, this is stats.norm.ppf(0.975)

# Calculate the Margin of Error (ME)
margin_of_error = z_critical * standard_error

# Calculate the Confidence Interval
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error

print(f"\n{confidence_level*100}% Confidence Interval for Population Proportion: ({lower_bound:.4f}, {upper_bound:.4f})")
```

### Coding Exercise: Automobile Insurance Premium

The average annual premium for automobile insurance in the United States is \\$1,503. The following annual premiums (\\$) are representative of the web site’s findings for the state of Michigan.

1905, 3112, 2312, 2725, 2545, 2981, 2677, 2525, 2627, 2600, 2370, 2857, 2962, 2545,	2675, 2184, 2529, 2115, 2332, 2442

Assume the population is approximately normal. 

1.	Provide a point estimate of the mean annual automobile insurance premium in Michigan.
1.	Develop a 95% confidence interval for the mean annual automobile insurance premium in Michigan.
1.	Does the 95% confidence interval for the annual automobile insurance premium in Michigan include the national average for the United States? What is your interpretation of the relationship between auto insurance premiums in Michigan and the national average?


```python
import numpy as np
from scipy import stats

premiums = [1905, 3112, 2312, 2725, 2545, 2981, 2677, 2525, 2627, 2600, 2370, 2857, 2962, 2545, 2675, 2184, 2529, 2115, 2332, 2442]
mean_premium = np.mean(premiums)

std_premium = np.std(premiums, ddof=1)

n = len(premiums)

sem = std_premium / np.sqrt(n)

t_value = stats.t.ppf(0.975, df=n-1)  # 0.975 for 95% CI
lower = mean_premium - t_value * sem
upper = mean_premium + t_value * sem

print("1. Point estimate of the mean annual automobile insurance premium in Michigan: ", round(mean_premium, 2))
print("2. Mean annual automobile insurance premium in Michigan: ", round(mean_premium, 2))
print("3. 95% confidence interval for mean: ", round(lower, 2), "to", round(upper, 2))

national_avg = 1503
if lower <= national_avg <= upper:
    print("3a. The national average IS inside the confidence interval.")
else:
    print("3a. The national average is NOT inside the confidence interval.")
```

### Coding Exercise: Will Our Children Be Better Off?

One of the questions on a survey of 1,000 adults asked if today’s children will be better off than their parents. Representative data are shown in the file _06_childoutlook.csv_. A response of Yes indicates that the adult surveyed did think today’s children will be better off than their parents. A response of No indicates that the adult surveyed did not think today’s children will be better off than their parents. Not Sure is the third option for those adults who are unsure about the answer.

1.	What is the point estimate of the proportion of the population of adults who do think that today’s children will be better off than their parents?
1.	At 95% confidence, what is the margin of error?
1.	What is the 95% confidence interval for the proportion of adults who do think that today’s children will be better off than their parents?
1.	What is the 95% confidence interval for the proportion of adults who do not think that today’s children will be better off than their parents?
1.	Which of the confidence intervals in parts (c) and (d) has the smaller margin of error? Why?


```python
import pandas as pd
import numpy as np
from scipy import stats

survey_data = pd.read_csv("C://Users//marin//Downloads//06_childoutlook.csv", delimiter='\t')

responses = survey_data.iloc[:, 0].str.strip().str.lower()

sample_size = len(responses)

yes_responses = (responses == "yes").sum()
no_responses = (responses == "no").sum()

prop_yes = yes_responses / sample_size
prop_no = no_responses / sample_size

z_critical = stats.norm.ppf(0.975)

se_yes = np.sqrt(prop_yes * (1 - prop_yes) / sample_size)
me_yes = z_critical * se_yes

se_no = np.sqrt(prop_no * (1 - prop_no) / sample_size)
me_no = z_critical * se_no

print("1. The point estimate of the proportion of the population of adults who do think that today’s children will be better off than their parents is:", round(prop_yes, 4))
print("2. The margin of error at 95% confidence for this estimate is:", round(me_yes, 4))
print("3. The 95% confidence interval for the proportion of adults who do think that today’s children will be better off than their parents is:", round(prop_yes - me_yes, 4), "to", round(prop_yes + me_yes, 4))
print("4. The 95% confidence interval for the proportion of adults who do not think that today’s children will be better off than their parents is:", round(prop_no - me_no, 4), "to", round(prop_no + me_no, 4))

if me_yes < me_no:
    print("5. The 'Yes' interval is narrower because the margin of error is smaller when the sample proportion is farther from 0.5.")
else:
    print("5. The 'No' interval is narrower because the margin of error is smaller when the sample proportion is farther from 0.5.")
```

### Manual Calculation Exercise: Company Profit

According to Thomson Financial, last year the majority of companies reporting profits had beaten estimates. A sample of 162 companies showed that 104 beat estimates, 29 matched estimates, and 29 fell short. 

1.	What is the point estimate of the proportion that fell short of estimates?
1.	Determine the margin of error and provide a 95% confidence interval for the proportion that beat estimates.
1.	How large a sample is needed if the desired margin of error is 0.05? __Hint:__ Reverse the formula for margin of error

Please manually calculate this problem. Show your process in the following box.

```python
import numpy as np
from scipy import stats

sample_size = 162
beat_companies = 104
matched_companies = 29
short_companies = 29

prop_short = short_companies / sample_size

prop_beat = beat_companies / sample_size
standard_error = np.sqrt(prop_beat * (1 - prop_beat) / sample_size)
z_critical = stats.norm.ppf(0.975)
margin_error = z_critical * standard_error
ci_beat_low = prop_beat - margin_error
ci_beat_high = prop_beat + margin_error

desired_margin_error = 0.05
required_sample_size = (z_critical**2 * prop_beat * (1 - prop_beat)) / (desired_margin_error**2)

print("1. The point estimate of the proportion of companies that fell short of estimates is:", round(prop_short, 4))
print("2. The 95% confidence interval for the proportion of companies that beat estimates is:", round(ci_beat_low, 4), "to", round(ci_beat_high, 4))
print("3. The required sample size to achieve a margin of error of 0.05 is:", int(np.ceil(required_sample_size)))
```

### Discussion Question: Store Sales

You are a regional manager for a national retail chain. To support staffing and inventory planning, you want to estimate the average weekly sales per store across your region. You collect data from a random sample of 10 stores:

Sample mean weekly sales: $\bar{x}$ = \\$24,000
<br>
Sample standard deviation: 𝑠 = \\$4,000

1. In this case, can we apply the Central Limit Theorem to compute the confidence interval? If we do, what assumptions are we making?
1. What would make this estimate more reliable?
1. How could this analysis mislead decision-makers if misunderstood?

```python
import numpy as np
from scipy import stats

# Given data
mean_sales = 24000
std_sales = 4000
n = 10

sem = std_sales / np.sqrt(n)

t_value = stats.t.ppf(0.975, df=n-1)
lower = mean_sales - t_value * sem
upper = mean_sales + t_value * sem

print("Problem: Store Sales")
print("1. 95% confidence interval for mean weekly sales:", round(lower,2), "to", round(upper,2))

print("\n2. Can we apply CLT?")
print("   - With n=10, CLT is weak. We assume sales are normally distributed.")
print("   - Also assume random sampling and independence.")

print("\n3. What would make this estimate more reliable?")
print("   - Larger sample size (more stores).")
print("   - Truly random and representative sample.")

print("\n4. How could this analysis mislead?")
print("   - CI is about the average, not a single store.")
print("   - Small n=10 means results can be unstable.")
print("   - If sample not representative, CI won’t reflect reality.")
```
